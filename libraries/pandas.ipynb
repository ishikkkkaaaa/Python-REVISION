```python
import matplotlib.pyplot as plt
```

```python
GDP=[25,35,45,55,65]
years=[2000,2001,2002,2003,2004]
plt.plot(years,GDP)
plt.xlabel("years")
plt.ylabel("GDP")
plt.scatter(years,GDP)
plt.text(2000,45,"low latency")
plt.grid(True)
plt.show()
```

```python
year=[2019,2020,2021]
cgpa=[8.05,8.81,9.08]
plt.plot(year,cgpa)
plt.xlabel("Years of College")
plt.ylabel("CGPA scored")
plt.scatter(year,cgpa,alpha=1)
plt.grid(True)
plt.show()
```

```python
teams=['groupA', 'groupB', 'groupC']
values=[10,20,50]
plt.figure(figsize=(10,3))
plt.subplot(131)
plt.bar(teams,values)
plt.subplot(132)
plt.plot(teams,values,linewidth=5)
plt.subplot(133)
plt.scatter(teams,values)
plt.suptitle('Categorical plotting')
plt.show()
```
