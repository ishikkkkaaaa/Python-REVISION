# MULTIPLE REGRESSION

```python
import pandas as pd
import numpy as np
```

```python
dataset=pd.read_csv('Concrete_Data.csv')
dataset.head()
```

```python
#createing x and y variable for dependent and independent variable col!
#x-> independet
#y-> dependent variable
x=dataset.drop(['Cement'],axis=1).values
y=dataset['Cement'].values
```

```python
print(x)
```

```python
print(y)
```

```python
#spliting dataset in training set and set test
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.33,random_state=0)
```

```python
#Train model on training set
from sklearn.linear_model import LinearRegression
ml=LinearRegression()
ml.fit(x_train,y_train)
```

#predict test set results(accuracy)
y_pred=ml.predict(x_test)
print(y_pred)

```python
ml.predict([[0.0,0.0,162.0,2.5,1040.0,676.0,28,79.99]])
```

```python
#evaulate the model
from sklearn.metrics import r2_score
r2_score(y_test,y_pred)
```

```python
#plot the result
import matplotlib.pyplot as plt
plt.figure(figsize=(20,15))
plt.scatter(y_test,y_pred,color='red')
plt.xlabel('actaul')
plt.xlabel('predicted')
plt.show()
```
