```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
boston_df = pd.read_csv('Boston.csv')
```

```python
plt.plot(boston_df['age'])
```

```python
plt.figure(figsize=(10,6))
plt.plot(boston_df['age'])
plt.title('Line plot for Age column')
plt.xlabel('Indices')
plt.ylabel('Age')
plt.show()
```

```python
X = [1,2,5,7,9,11]
Y = [10,15,12,32,45,76]
```

```python
plt.bar(X,Y)
```

```python
X1 = [1,2,5,7,9,11]
Y1= [10,15,12,32,45,76]
X2 = [2,4,6,8,10,12]
Y2 = [32,31,52,44,95,78]
```

```python
#histogram
plt.figure(figsize=(10,8))
plt.hist(boston_df['medv'])
plt.title('Histogram of MedValue(medv)')
plt.xlabel('medv range',fontsize = 12)
plt.ylabel('Frequency',fontsize = 12)
```

```python
#Scatter plot
iris_df = pd.read_csv('iris.csv')
iris_df.head()
iris_df['variety'].unique()
```

```python
plt.figure(figsize=(8,6))
plt.scatter(x=iris_df['sepal.length'],y=iris_df['sepal.width'],marker = '*')
```

```python
#seaborn
plt.figure(figsize=(8,6))
sns.scatterplot(x='sepal.length',y='sepal.width',data=iris_df,hue="sepal.width")
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.displot(boston_df['medv'],color='darkviolet')
plt.show()
```
