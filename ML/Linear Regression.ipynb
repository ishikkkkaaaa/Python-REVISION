```python
# Making imports
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = (20.0, 10.0)

# Preprocessing Input data
data = pd.read_csv('salary_data.csv')
X = data.iloc[:, 0]
Y = data.iloc[:, 1]
plt.scatter(X, Y)
plt.show()
```

```python
# Building the model
X_mean = np.mean(X)
Y_mean = np.mean(Y)

num = 0
den = 0
for i in range(len(X)):
    num += (X[i] - X_mean)*(Y[i] - Y_mean)
    den += (X[i] - X_mean)**2
m = num / den
c = Y_mean - m*X_mean

print(m, c)

```

```python
# Making predictions
Y_pred = m*X + c
plt.scatter(X, Y) 
# actual
plt.plot([min(X), max(X)], [min(Y_pred), max(Y_pred)], color='red') 
# predicted
plt.show()

```
